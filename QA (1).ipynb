```python
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import DistilBertForQuestionAnswering, DistilBertTokenizerFast, AdamW
from sklearn.model_selection import train_test_split
from tqdm import tqdm

class QADataset(Dataset):
    def __init__(self, questions, contexts, answers, tokenizer, max_length=384):
        self.encodings = tokenizer(
            questions,
            contexts,
            truncation=True,
            padding=True,
            max_length=max_length,
            return_tensors='pt',
            return_offsets_mapping=True
        )

        self.start_positions = [] # Prepare start and end positions
        self.end_positions = []

        for i, (question, context, answer) in enumerate(zip(questions, contexts, answers)):
            answer_start = context.find(answer)  # Find the start and end character positions of the answer in the context
            if answer_start == -1:
                self.start_positions.append(0)
                self.end_positions.append(0)
                continue

            answer_end = answer_start + len(answer)
            token_start = None
            token_end = None

            for j, (start, end) in enumerate(self.encodings.offset_mapping[i]):
                if start <= answer_start < end:
                    token_start = j
                if start < answer_end <= end:
                    token_end = j
                    break

            if token_start is not None and token_end is not None:
                self.start_positions.append(token_start)
                self.end_positions.append(token_end)
            else:
                self.start_positions.append(0)
                self.end_positions.append(0)

        self.start_positions = torch.tensor(self.start_positions)
        self.end_positions = torch.tensor(self.end_positions)

    def __len__(self):
        return len(self.encodings['input_ids'])

    def __getitem__(self, idx):
        item = {
            'input_ids': self.encodings['input_ids'][idx],
            'attention_mask': self.encodings['attention_mask'][idx],
            'start_positions': self.start_positions[idx],
            'end_positions': self.end_positions[idx]
        }
        return item

def finetune_distilbert(data_path, model_name='distilbert-base-uncased-distilled-squad', epochs=3, batch_size=16, learning_rate=5e-5):

    df = pd.read_csv(data_path)
    df = df.dropna()

    train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)

    tokenizer = DistilBertTokenizerFast.from_pretrained(model_name)
    model = DistilBertForQuestionAnswering.from_pretrained(model_name)

    train_dataset = QADataset(
        train_df['question'].tolist(),
        train_df['context'].tolist(),
        train_df['answer'].tolist(),
        tokenizer
    )
    val_dataset = QADataset(
        val_df['question'].tolist(),
        val_df['context'].tolist(),
        val_df['answer'].tolist(),
        tokenizer
    )

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size)

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.to(device)
    optimizer = AdamW(model.parameters(), lr=learning_rate)

    for epoch in range(epochs):
        model.train()
        total_train_loss = 0

        for batch in tqdm(train_loader, desc=f'Epoch {epoch+1}/{epochs}'):
            # Move batch to device
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            start_positions = batch['start_positions'].to(device)
            end_positions = batch['end_positions'].to(device)

            optimizer.zero_grad()

            outputs = model(
                input_ids=input_ids,
                attention_mask=attention_mask,
                start_positions=start_positions,
                end_positions=end_positions
            )
            loss = outputs.loss
            total_train_loss += loss.item()

            loss.backward()
            optimizer.step()

        model.eval() # Validation
        total_val_loss = 0

        with torch.no_grad():
            for batch in tqdm(val_loader, desc='Validation'):
                input_ids = batch['input_ids'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                start_positions = batch['start_positions'].to(device)
                end_positions = batch['end_positions'].to(device)

                outputs = model(
                    input_ids=input_ids,
                    attention_mask=attention_mask,
                    start_positions=start_positions,
                    end_positions=end_positions
                )
                loss = outputs.loss
                total_val_loss += loss.item()

        avg_train_loss = total_train_loss / len(train_loader)
        avg_val_loss = total_val_loss / len(val_loader)
        print(f'Epoch {epoch+1}/{epochs}')
        print(f'Average training loss: {avg_train_loss:.4f}')
        print(f'Average validation loss: {avg_val_loss:.4f}')

    model.save_pretrained('./finetuned_distilbert_qa')
    tokenizer.save_pretrained('./finetuned_distilbert_qa')

    return model, tokenizer


def predict_answer(model, tokenizer, question, context, max_length=384):
    # Prepare inputs
    inputs = tokenizer(
        question,
        context,
        return_tensors='pt',
        truncation=True,
        max_length=max_length
    )

    # Get model prediction
    with torch.no_grad():
        outputs = model(**inputs)

    # Get the most likely answer span
    start_logits, end_logits = outputs.start_logits, outputs.end_logits
    start_index = torch.argmax(start_logits)
    end_index = torch.argmax(end_logits)

    # Decode the answer
    answer_tokens = inputs['input_ids'][0][start_index:end_index+1]
    answer = tokenizer.decode(answer_tokens)

    return answer
```

```python
import torch
import pandas as pd
from transformers import DistilBertForQuestionAnswering, DistilBertTokenizerFast

def prepare_test_data(test_csv_path):

    test_df = pd.read_csv(test_csv_path)
    test_df = test_df.dropna()
    test_df = test_df.sample(frac=0.2, random_state=64)

    required_columns = ['question', 'context', 'answer']
    for col in required_columns:
        if col not in test_df.columns:
            raise ValueError(f"Missing required column: {col}")

    return (
        test_df['question'].tolist(),
        test_df['context'].tolist(),
        test_df['answer'].tolist()
    )

def compute_metrics(predictions, ground_truth):

    def normalize_text(text):
        return ' '.join(str(text).lower().split())

    def compute_f1(prediction, truth):
        pred_tokens = set(normalize_text(prediction).split())
        true_tokens = set(normalize_text(truth).split())
        common_tokens = pred_tokens.intersection(true_tokens)

        if len(pred_tokens) == 0 or len(true_tokens) == 0:
            return 0

        precision = len(common_tokens) / len(pred_tokens)
        recall = len(common_tokens) / len(true_tokens)

        if precision + recall == 0:
            return 0
        return 2 * (precision * recall) / (precision + recall)

    exact_matches = []
    f1_scores = []

    for pred, truth in zip(predictions, ground_truth):
        exact_matches.append(1 if normalize_text(pred) == normalize_text(truth) else 0)
        f1_scores.append(compute_f1(pred, truth))

    import numpy as np
    metrics = {
        'exact_match': np.mean(exact_matches),
        'f1_score': np.mean(f1_scores),
        'precision': len([x for x in exact_matches if x]) / len(exact_matches),
        'recall': len([x for x in f1_scores if x > 0.5]) / len(f1_scores)
    }

    return metrics

def evaluate_model(test_csv_path, model_path='./finetuned_distilbert_qa', max_length=384):

    questions, contexts, ground_truth = prepare_test_data(test_csv_path)

    model = DistilBertForQuestionAnswering.from_pretrained(model_path)
    tokenizer = DistilBertTokenizerFast.from_pretrained(model_path)

    model.eval()

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.to(device)

    # Predict answers
    predictions = []
    for question, context in zip(questions, contexts):
        # Prepare inputs
        inputs = tokenizer(
            question,
            context,
            return_tensors='pt',
            truncation=True,
            max_length=max_length
        ).to(device)

        with torch.no_grad():
            outputs = model(**inputs)


        start_logits, end_logits = outputs.start_logits, outputs.end_logits # Get the most likely answer span
        start_index = torch.argmax(start_logits)
        end_index = torch.argmax(end_logits)

        # Decode the answer
        answer_tokens = inputs['input_ids'][0][start_index:end_index+1]
        prediction = tokenizer.decode(answer_tokens)
        predictions.append(prediction)

    # Compute metrics
    metrics = compute_metrics(predictions, ground_truth)

    # Detailed results
    detailed_results = pd.DataFrame({
        'Question': questions,
        'Context': contexts,
        'Ground Truth': ground_truth,
        'Prediction': predictions
    })

    # Error analysis
    detailed_results['Is_Correct'] = detailed_results.apply(
        lambda row: row['Ground Truth'].lower().strip() == row['Prediction'].lower().strip(),
        axis=1
    )

    return {
        'metrics': metrics,
        'detailed_results': detailed_results,
        'incorrect_predictions': detailed_results[~detailed_results['Is_Correct']]
    }

def main():

    results = evaluate_model('qa_dataset_test.csv')

    print("Evaluation Metrics:")
    for metric, value in results['metrics'].items():
        print(f"{metric.replace('_', ' ').title()}: {value:.4f}")

    results['detailed_results'].to_csv('qa_test_results.csv', index=False)
    results['incorrect_predictions'].to_csv('qa_incorrect_predictions.csv', index=False)

    print("\nSample Incorrect Predictions:")
    print(results['incorrect_predictions'].head())

if __name__ == '__main__':
    main()
```
